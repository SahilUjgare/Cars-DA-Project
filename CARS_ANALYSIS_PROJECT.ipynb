#  Introduction

The automobile industry is one of the fastest-growing markets worldwide, with a wide variety of vehicles catering to different customer needs. Understanding the factors that influence **car prices** is essential not only for buyers and sellers but also for manufacturers, dealers, and financial institutions. This project aims to explore and analyze a dataset containing information about **car brands, models, manufacturing year, engine size, fuel type, transmission type, mileage, doors, ownership history, and price**.

The primary objective of this project is to perform **Exploratory Data Analysis (EDA)** to identify meaningful patterns, trends, and relationships among these features. By addressing key data-driven questions, the analysis highlights how different factors such as **brand reputation, mileage, fuel type, transmission, and age** affect the pricing of cars in the resale market.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Load the file (.csv)
cars_df = pd.read_csv("/content/car_price_dataset.csv")
```

```python
cars_df.head()
```

```python
#1.What are the top 10 most common car brands in the dataset?
most_common_brands = cars_df['Brand'].value_counts().sort_index()
most_common_brands
```

```python
# Create a bar plot for the most common car brands
plt.figure(figsize=(10, 6))
sns.barplot(x=most_common_brands.index, y=most_common_brands.values,palette = "magma")
plt.title('Top 10 Most Common Car Brands')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#2.What is the average price of a car by brand?
avg_price = cars_df.groupby('Brand')['Price'].mean()
display(avg_price)
```

```python
# Create a bar plot for the average price by brand
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price.index, y=avg_price.values, palette="viridis")
plt.title('Average Car Price by Brand')
plt.xlabel('Brand')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#3.Which year has the highest number of cars listed?
highest_number_of_cars = cars_df.groupby('Brand')['Year'].value_counts()
highest_number_of_cars = highest_number_of_cars.idxmax()
highest_number_of_cars
```

```python
#4.What is the distribution of fuel types across all cars?
fuel_types = cars_df.groupby('Brand')['Fuel_Type'].value_counts()
fuel_types
```

```python
# Create a bar plot for the distribution of fuel types
plt.figure(figsize=(12, 6))
sns.countplot(data=cars_df, x='Fuel_Type', palette='icefire')
plt.title('Distribution of Fuel Types')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#5.How does the average car price vary by transmission type (Manual, Automatic, Semi-Automatic)?
avg_price_by_transmission = cars_df.groupby('Transmission')['Price'].mean()
display(avg_price_by_transmission)
```

```python
# Create a bar plot for the average price by transmission type
plt.figure(figsize=(8, 5))
sns.barplot(x=avg_price_by_transmission.index, y=avg_price_by_transmission.values, palette='rocket')
plt.title('Average Car Price by Transmission Type')
plt.xlabel('Transmission Type')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
#6.What is the minimum, maximum, and average mileage in the dataset?
avg_mileage = cars_df.groupby('Brand')['Mileage'].mean()
avg_mileage
```

#  Conclusion

The **Car Price Dataset Analysis Project** provided valuable insights into the factors influencing car prices, mileage, and overall market distribution. Through **Exploratory Data Analysis (EDA)** and formulation of key data-driven questions, several important patterns and relationships were uncovered:

1. **Brand Influence**

   * Certain brands dominate the dataset in terms of count and popularity.
   * Luxury brands (e.g., Mercedes, Audi, BMW) tend to have higher average prices, whereas brands like Kia, Hyundai, and Chevrolet represent more budget-friendly options.

2. **Year of Manufacture & Depreciation**

   * Cars manufactured in recent years (after 2020) generally have **higher average prices** compared to older cars.
   * A clear **depreciation trend** was observed, where the resale value drops significantly as cars age, but some premium brands retain value better.

3. **Fuel Type Trends**

   * **Petrol and Diesel cars** dominate the dataset, with Diesel often associated with higher mileage.
   * **Electric cars** are fewer but generally priced higher due to modern technology and lower operating costs.
   * **Hybrid cars** show mixed pricing patterns, often influenced by engine size and brand positioning.

4. **Transmission and Ownership**

   * **Automatic cars** usually have a higher average price compared to manual ones, reflecting modern consumer demand.
   * Cars with more owners tend to have lower prices, showing that ownership history significantly affects resale value.

5. **Mileage vs. Price Relationship**

   * A **negative correlation** was found between mileage and price  higher mileage generally lowers car value.
   * However, exceptions exist where luxury cars maintain higher prices despite relatively higher mileage.

6. **Engine Size and Features**

   * Larger engine sizes tend to correlate with higher prices, but only up to a certain point (very large engines sometimes depreciate faster due to fuel inefficiency).
   * Features like **doors and transmission type** also slightly influence prices but are less impactful than brand, year, and mileage.

---

##  Key Takeaways

* **Car Price is a Multi-factor Outcome:** It depends on brand reputation, age, fuel type, mileage, transmission, and ownership history.
* **Depreciation is Inevitable:** Car prices drop as years pass, but premium brands hold value better.
* **Emerging Market Trends:** Electric and hybrid cars are slowly gaining presence, often with higher initial prices.
* **Mileage & Ownership History:** These are the strongest indicators of resale value alongside brand and year.
